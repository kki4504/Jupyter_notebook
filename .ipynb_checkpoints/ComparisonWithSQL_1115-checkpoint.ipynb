# 복사 vs. 원본변경 연산 (Copies vs. in place operations)

#### 대부분의 pandas 연산은 Series/DataFrame 복사본을 반환한다.
#### 변경 내용을 원본에 반영하기 위해서는 새로운 변수에 할당하거나
    -> sorted_df = df.sort_values('col1')  
#### 또는 원본 파일에 덮어쓴다.
    df = df.sort_values('col1')

Note : 일부 매서드들은 inplace=True 키워드 인자를 제공한다. df.sort_values('col1', inplace=True) 이런 방법은 권장되지 않는다.  

### SELECT

SQL에서 selection은 쉼표로 구분되는 칼럼 리스트를 이용한다. (모든 칼럼을 선택하기 위해서는 * 를 사용할 수 있다.)  
    - SELECT total_bill, tip, smoker, time  
    - FROM tips;

pandas에서는 칼럼 선택을 DataFrame에 칼럼 리스트를 전달함으로써 수행할 수 있다.

```python
tips[['total_bill', 'tip', 'smoker', 'time']]
```

```python
칼럼 이름 리스트 없이 DataFrame
```

```python
tips
```

SQL에서 계산된 칼럼을 추가할 수 있다.  
        SELECT *, tip/total_blii as tip_rate  
        FROME tips

```python
df = tips
```

```python
df
```

```python
df['tip_rate'] = df['tip'] / df['total_bill']
```

```python
df
```

```python
tips
```

```python
tips = tips.drop(columns=['tip_rate'])
```

```python
tips
```

위에처럼 해도 되고,  
pandas에서, DataFrame의 assign()메서드를 이용해 새로운 칼럼을 추가할 수 있다.

```python
tips.assign(tip_rate = tips['tip']/tips['total_bill'])
```

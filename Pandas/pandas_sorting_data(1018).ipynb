```python
import pandas as pd
```

DataFrame을 만들기 위해 파이썬 dictionary를 다음과 같이 정의한다.

```python
people = {
    "first" : ['Gildong', 'Chunhyang', 'Mongyoung'],
    "last" : ['Hong', 'Sung', 'Lee'],
    "email" : ['gdhong@gmail.com', "chsung@gmail.com", "mrlee@gmail.com"]
}
```

people로 부터 DataFrame을 생성하고, 그놈을 df라 하자.

```python
df = pd.DataFrame(people);
```

```python
df
```

```python
type(df)
```

last 칼럼값을 오름차순으로 정렬하자.  sort_values() 함수이용,

```python
df.sort_values(by='last')
```

원본은 변경되지 않았고, 오름차순 정렬이 됐다.

```python
df
```

기본적으로 sort_values()는 오름차순 정렬한다. sort_vlaues(ascending = True)가 기본값.  
내림차순으로 정렬하고자 한다면 sort_values(ascending=False)

```python
df.sort_values(by="last", ascending=False)
```

SQL에서도 order by 절에 정렬 기준을 여러개 명시할 수 잇다.  
pandas도 그렇다

```python
df.sort_values(by = ["last", "first"], ascending=False)
```

last 는 오름차순 first는 내림차순

```python
df.sort_values(by = ['last', 'first'], ascending=[ True, False])
```

연봉 상위 10명 출력

pandas에서 상위 n개의 결과를 볼 수 있는 nlargest() 함수를 제공한다.  
상위 10명의 연봉이 보고싶다면 : 

```python
df['SalaryUSD'].nlargest(10)
```

연봉 top 10의 다른 정보도 알고싶다 그러면?

```python
df.nlargest(10, 'SalaryUSD')[['Country', 'EdLevel','SalaryUSD']]
```

최소연봉 top10을 알고싶다면 nsamllest()

```python
df.nsmallest(10, 'SalaryUSD')[['Country', 'EdLevel', 'SalaryUSD']]
```

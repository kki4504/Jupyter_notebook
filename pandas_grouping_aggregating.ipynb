# 각 나라의 응답자 중 몇 %가 파이썬을 사용하나?

각 나라별로  
   - 파이썬을 사용하는 응답자 수 / 응답자 수

```python
df['Country'].value_counts()
```

```python
country_respondents = df['Country'].value_counts()
```

```python
country_respondents
```

```python
country_python_uses =  country_grp['LanguageWorkedWith'].apply(lambda x : x.str.contains('Python').sum())
```

pandas의 concat 함수를 이용해서 여러 시리즈를 조합할 수 있다

```python
concat_df = pd.concat([country_respondents, country_python_uses])
```

```python
concat_df
```

```python
concat_df = pd.concat([country_respondents, country_python_uses], axis='columns')
```

```python
concat_df
```

concat_df.rename(columns={'origianl colname1' : 'new colname1', 'original colname2': 'new colname2'})

```python
concat_df.rename(columns={'Country' : 'NumRespondents' , 'LanguageWorkedWith' : 'NumKnowsPython'})
```

```python
concat_df.rename(columns={'Country' : 'NumRespondents' , 'LanguageWorkedWith' : 'NumKnowsPython'}, inplace=True)
```

```python
concat_df
```

```python
concat_df['PercentageKnowPython'] = (concat_df['NumKnowsPython'] / concat_df['NumRespondents'])
```

```python
concat_df
```

```python
concat_df.sort_values(by = 'PercentageKnowPython', ascending=False, inplace=True)
```

```python
concat_df
```

```python
concat_df.head(50)
```

# 응답자 중 학사학위 이상의 교육수준을 가진 ㅂㅣ율을 나라별로 구하시오.

```python
df['EdLevel'].unique()
```

각 나라별 학사학위는 Bachelor 라는 문자열을 포함하는 응답자

concat_df2 라는 DataFrame을 만들어라 ( 나라별 응답자 수 Bachelor 학위 소지자 수 , 학위 소지 비율) 로 구성

```python
country_bachelor = country_grp['EdLevel'].apply(lambda x : x.str.contains('Bachelor').sum())
```

```python
country_bachelor
```

```python
country_respondents
```

```python
concat_df2 = pd.concat([country_respondents, country_bachelor], axis='columns')
```

```python
concat_df2
```

```python
concat_df2.rename(columns={'Country' : 'NumRespondents', 'EdLevel' : 'NumBachelor'}, inplace=True)
```

```python
concat_df2
```

```python
concat_df2['PercentageBachelor'] = (concat_df2['NumBachelor'] / concat_df2['NumRespondents']) * 100
```

```python
concat_df2
```

```python
concat_df2.sort_values(by='PercentageBachelor', ascending=False, inplace=True)
```

```python
concat_df2
```

```python
concat_df2.nlargest(50, 'PercentageBachelor')
```
